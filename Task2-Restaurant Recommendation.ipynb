```python
df1= df.drop(columns=['Restaurant ID','Address', 'Locality Verbose','Locality','Rating color'])
```

```python
df1['Cuisines'] = df1['Cuisines'].str.replace(',' , '') 
df1['Cuisines'] = df1['Cuisines'].astype(str).apply(lambda x: ' '.join(sorted(x.split())))
df1['Cuisines'].value_counts().head()
```

```python
df1['Currency'] = df1['Currency'].str.replace(r'\(.*\)', '', regex=True)
```

```python

from sklearn.preprocessing import LabelEncoder

categorical_columns = [ 'Currency', 'Has Table booking', 
                         'Has Online delivery', 'Is delivering now', 'Switch to order menu','Rating text']
labelencoder = LabelEncoder()

for col in categorical_columns:
    df1[col] = labelencoder.fit_transform(df1[col])

df1.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df1['Cuisines'])


cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python

def get_recommendations(user_preferences, top_n=5):
    indices = df.index[df['Cuisines'] == user_preferences['Cuisine']].tolist()
    indices = [idx for idx in indices if df.loc[idx, 'Price range'] == user_preferences['Price range']]
    if not indices:
        print("No restaurants found matching the given preferences.")
        return []
    sim_scores = [(idx, 1.0) for idx in indices] 
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[:top_n]
    restaurant_indices = [x[0] for x in sim_scores]
    return df.iloc[restaurant_indices]['Restaurant Name']

user_preferences = {'Cuisine': 'Chinese', 'Price range': 2}  

recommendations = get_recommendations(user_preferences)
print(recommendations)
```

#### RECOMMENDATION SYSYTEM BY PRICE RANGE,RATING,HAS ONLINE DELIVERY, CUISINES

```python

def get_recommendations(user_preferences, top_n=5):
    
    indices = df.index[df['Cuisines'] == user_preferences['Cuisine']].tolist()
    
    indices = [idx for idx in indices if df.loc[idx, 'Price range'] == user_preferences['Price range']]
   
    indices = [idx for idx in indices if df.loc[idx, 'Aggregate rating'] >= user_preferences.get('Min Rating', 0)]
    
    if user_preferences.get('Has Online delivery', False):
        indices = [idx for idx in indices if df.loc[idx, 'Has Online delivery'] == 'Yes']
    if not indices:
        print("No restaurants found matching the given preferences.")
        return []
    
    sim_scores = [(idx, 1.0) for idx in indices]  
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[:top_n]
    
    restaurant_indices = [x[0] for x in sim_scores]
    
    return df.iloc[restaurant_indices]['Restaurant Name']


user_preferences = {'Cuisine': 'Chinese', 'Price range': 2, 'Min Rating': 4.0, 'Has Online delivery': True}


recommendations = get_recommendations(user_preferences)
print(recommendations)
```
